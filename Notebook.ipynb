# Training

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

Selecting variable names as x_train,x_test,y_train,y_test and pushing the values from previous datasets. 
/// train_test_split(x,y,test_size=0.3) means train data will be splitted and test part will be .3(30%) so remaining is the train data 0.7(70%)

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

Imported Random Forest. You can press shift+tab in between r( to see parameters. // before starting () you can check parameters if you dont know

```python
## Hyperparameter
import numpy as np
n_estimators=[int(x) for x in np.linspace(start=100, stop =1200,num=12)]
print(n_estimators)
```

n_estimators is number of trees in random forest

```python
max_features  = ['auto','sqrt']
```

number of features to consider at every split

```python
max_depth=[int(x) for x in np.linspace(5,30,num=6)]
```

number of levels

```python
min_samples_split =[2,5,10,15,100]
```

minimum number of sample to split a node

```python
min_samples_leaf = [1,2,5,10]
```

minimum number of samples require at each leaf node

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = { 'n_estimators':n_estimators,
               'max_features':max_features,
               'max_depth':max_depth,
               'min_samples_split':min_samples_split,
               'min_samples_leaf':min_samples_leaf
    
}
```

```python
print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(x_train,y_train)
```

# Predictions

```python
predictions=rf_random.predict(x_test)
```

```python
predictions
```

```python
sns.displot(y_test-predictions)
```
